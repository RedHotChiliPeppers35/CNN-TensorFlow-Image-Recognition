```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras import layers, models
import cv2
import os
import splitfolders
```

```python
input_folder = "/Users/ataberkcinetci/Desktop/Agricultural-crops"
output_folder = "/Users/ataberkcinetci/Desktop/output_file"

split_ratio = (0.8, 0.1, 0.1)

splitfolders.ratio(
    input_folder,
    output=output_folder,
    seed=500,
    ratio=split_ratio,
    group_prefix=None,  
)

```

```python
img_size=(224,224)
batch_size=32

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest",

)    
```

```python
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
valid_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
train_dir=os.path.join(output_folder,"train")
val_dir=os.path.join(output_folder,"val")
test_dir=os.path.join(output_folder,"test")


train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
)

test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
)

val_data = valid_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
)

```

```python

import random
images, labels = next(val_data)
idx = random.randint(0,images.shape[0]-1)
```

```python
from keras.applications.resnet50 import ResNet50
base_model = ResNet50(weights= "imagenet",include_top=False,input_shape=(img_size[0], img_size[1], 3))
base_model.trainable= False
```

```python
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(30,activation="softmax"),
])
```

```python
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(train_data,epochs=25, validation_data=val_data)
```

```python
test_loss, test_accuracy = model.evaluate(test_data)
print(test_accuracy)
```
